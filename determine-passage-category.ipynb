# Buffer Analysis
## Description 
Count GPS points within buffers around collection sites
to determine what drives high/low passage classification.

## Steps
- creating buffers around collection sites
- overlapping data points from different origins to understand low/ high passage classes:
    - subcohort or GPS follow-up or some combination or all
    - varying buffer sizes
- saves median split results: which threshold can explain the split, specific to each LU type

## Constraints
- All villages location were considered 'high passage' regardless of the gps data points, so they are not included in the analysis. 
- The analysis was executed on a subset of data, including only the first round of mosquito collection and HDN locations.
